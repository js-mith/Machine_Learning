```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words=[]
    for w in  words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word =lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [ (clean_review(document) , category) for document,category in document]
```

```python
documents[0]
```

```python
training_document = documents[0:1500]
testing_document = documents[1500:]
```

```python
all_word = []
for doc in training_document:
    all_word+=doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_word)
common = freq.most_common(3000)
feature = [i[0] for i in common ]
```

```python
feature
```

```python
def get_feature_dict(words): #will return true/false if the word in present in the document or not
    current_features = {}
    words_set = set(words)
    for w in feature:
        current_features[w] = w in words_set
    return current_features
     

```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in training_document] 
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_document]
    
```

```python
#Classification using NLTK Naive Bayes   
from nltk import NaiveBayesClassifier 
classifier = NaiveBayesClassifier.train(training_data)
nltk.classify.accuracy(classifier, testing_data)
```

```python
classifier.show_most_informative_features(15)
```
