```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words=[]
    for w in  words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word =lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [ (clean_review(document) , category) for document,category in document]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
train_set = {"the sky sky is blue " , "the sun is bright"}
count_vec = CountVectorizer(max_features =3)
a = count_vec.fit_transform(train_set)
a.todense()
```

```python
categories = [category for document , category in documents]
```

```python
text_document =[''.join(document) for document,category in documents]
```

```python
from sklearn.model_selection import train_test_split
```
